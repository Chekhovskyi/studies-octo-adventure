## Знаходження значення визначеного інтеграла за квадратурними формулами, і формулою Сімпсона при розбитті на 10, 100 і 1000 відрзіків

```python
def f_28(x):
    return exp(x) * sqrt(x)
a = 5
b = 10
f = f_28 

x_etalone = arange(a, b + 0.00001, 0.00001)
y_etalone = [f(v) for v in x_etalone]
tr_etalone = trapz(y=y_etalone, x=x_etalone)

segments = [10, 100, 1000]
xs = [linspace(a, b, s) for s in segments]
ys = [[f(v) for v in x] for x in xs]

les = [left_endpoint(x, ys[i], f, a, b) for (i, x) in enumerate(xs)]
res = [right_endpoint(x, ys[i], f, a, b) for (i, x) in enumerate(xs)]
trs = [trapezoidal(x, ys[i], f) for (i, x) in enumerate(xs)]
sis = [simpson(x, ys[i], f) for (i, x) in enumerate(xs)]

display_table(["", "10", "100", "1000"],[
    ["Ліві прямокутники"] + les,
    ["Ліві прямокутники(АП)"] + abs_error(les, tr_etalone),
    ["Ліві прямокутники(ВП)"] + rel_error(les, tr_etalone),
    ["Праві прямокутники"] + res,
    ["Праві прямокутники(АП)"] + abs_error(res, tr_etalone),
    ["Праві прямокутники(ВП)"] + rel_error(res, tr_etalone),
    ["Tрапецій"] + trs,
    ["Tрапецій(АП)"] + abs_error(trs, tr_etalone),
    ["Tрапецій(ВП)"] + rel_error(trs, tr_etalone),
    ["Сімпсона"] + sis,
    ["Сімпсона(АП)"] + abs_error(sis, tr_etalone),
    ["Сімпсона(ВП)"] + rel_error(sis, tr_etalone)
])

figure(1)
title("Результати")
plot(segments, les)
plot(segments, res)
plot(segments, trs)
plot(segments, sis)
legend(["Ліві прямокутники", "Праві прямокутники", "Tрапецій", "Сімпсона"])
figure(2)
title("Абсолютна похибка")
plot(segments, abs_error(les, tr_etalone))
plot(segments, abs_error(res, tr_etalone))
plot(segments, abs_error(trs, tr_etalone))
plot(segments, abs_error(sis, tr_etalone))
legend(["Ліві прямокутники", "Праві прямокутники", "Tрапецій", "Сімпсона"])
figure(3)
title("Відносна похибка")
plot(segments, rel_error(les, tr_etalone))
plot(segments, rel_error(res, tr_etalone))
plot(segments, rel_error(trs, tr_etalone))
plot(segments, rel_error(sis, tr_etalone))
legend(["Ліві прямокутники", "Праві прямокутники", "Tрапецій", "Сімпсона"])
```
