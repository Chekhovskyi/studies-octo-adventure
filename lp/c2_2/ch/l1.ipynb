```python
%matplotlib inline
from pylab import *
from numpy import *
from utils import *
#from l1 import *
```

```python
N = [11, 121, 901]
A = [random.rand(n) for n in N]
```

```python
# Оцінити похибку обмеження при обчисленні функції розкладом у ряд
```

```python
# Оцінити похибку заокруглення
Sd = [sum(a) for a in A] # Сумма подвійної точності
Ss = [sum(float32(a)) for a in A] # Сумма одинарної точності
Ssa = [sum(float32(sorted(a))) for a in A] # Сумма сорованих
Ssd = [sum(float32(sorted(a, reverse=True))) for a in A] # Сумма сорованих спадаючи

Ead = abs_error(Sd, Ss)
Easa = abs_e(Sd, Ssa))
Easd = abs(subtract(Sd, Ssd))

Vad = abs(divide(Sd, Ss))
Vasa = abs(divide(Sd, Ssa))
Vasd = abs(divide(Sd, Ssd))

# Print results

display_table(["", "N1", "N2", "N3"],[
    ["Подвійна до одинарної(АП)"] + list(Ead),
    ["До сортованної(АП)"] + list(Easa),
    ["До обратно сортованної(АП)"] + list(Easd),
    ["Подвійна до одинарної(ВП)"] + list(Vad),
    ["До сортованної(ВП)"] + list(Vasa),
    ["До обратно сортованної(ВП)"] + list(Vasd)
])

figure(1)
title("Абсолютна похибка")
plot(N, Ead)
plot(N, Easa)
plot(N, Easd)
legend(["Подвійна до одинарної", "До сортованної", "До обратно сортованної"])
figure(2)
title("Відносна похибка")
plot(N, Vad)
plot(N, Vasa)
plot(N, Vasd)
legend(["Подвійна до одинарної", "До сортованної", "До обратно сортованної"])
```
