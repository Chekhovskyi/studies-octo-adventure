## Завдання 2
Побудувати графік функції наведеної у додатку 2 згідно варіанту індивідуального завдання отриманого у викладача. На заданому інтервалі для 5 різних точок побудувати інтерполяційний поліном (степінь див. додаток 2). За допомогою інтерполяційного полінома дослідити вплив степені полінома на точність одержаних результатів. Навести хід розв’язку. Написати програму
![](k_2_task.png)

Степіні полінома 4 і 7

```python
from interpolation import *
```

```python
f = lambda x: 0 if x < 2 else pow(x, 2)
n = 5 # Кількість точок

def experiment(a, b, d):
    x_all = arange(a-1, b+2, 0.1)
    plot(x_all, [f(xv) for xv in x_all])

    # Train
    x_train = linspace(a, b, d)
    l = lambda x_point: lagrange(x_point, x_train, f)
    plot(x_train, [f(xv) for xv in x_train])

    # Test
    x_test = linspace(a-1, b+2, n)
    plot(x_test, [l(xv) for xv in x_test])
    legend(["Оригінал", "В точках коеєфіцієнтів", "Інтерпольовані"])
```

#### Від -2 до 3, степінь полінома 4

```python
experiment(-2, 3, 4)
```

#### Від -2 до 3, степінь полінома 7

```python
experiment(-2, 3, 7)
```

#### Від 3 до 10, степінь полінома 4

```python
experiment(3,10, 4)
```

#### Від 3 до 10, степінь полінома 7

```python
experiment(3,10, 7)
```
