```python
%matplotlib inline
from pylab import *
from numpy import *
from utils import *
from l2 import *
```

```python
def f_27(x):
    return 1 / ( 1 + 5 * pow(x, 2) )
```

```python
# [-1;1]
f = f_27
start = -1
end = 1
degree = 8
x = linspace(start, end, 5)
y = [f(v) for v in x]

# Training
Tx = linspace(start, end, 9)
Ty = [f(v) for v in Tx]

plot(x, y)
plot(x, [newton1(Tx, Ty, v, degree) for v in x])
plot(x, [newton2(Tx, Ty, v, degree) for v in x])
legend(["Original", "Newton #1", "Newton #2"])
```
