```python
%matplotlib inline
import pandas as pd
from IPython.display import display

from l2_4_images import *
from l2_4_algs import *
```

```python
letters = "VGYJM≡Б457"

image_indexes = ["A" + str(i) for i in range(len(letters))]
base_images = list(gen_images(letters))

qualifier_masks = { 
   'S1' : sign_center_4x4_mask,
   'S2' : sign_line_y_7_mask,
   'S3' : sign_line_x_7_mask,
   'S4' : sign_diagonal_mask,
   'S5' : sign_all_pixels_mask
}
```

#### 5.5 розроблену систему ознак та параметрів методів (у вигляді табл. 4.2)

```python
Table(["#", "Графічна інтерпритація"],
    [
        [key, bw_matrix_from_mask(qualifier_masks[key])] 
        for key in sorted(qualifier_masks)
    ]
)
```

#### 5.6 значення ознак для символів еталонного набору

```python
base_image_qalifiers = pd.DataFrame({
            key:[calc_sign(img, qualifier_masks[key]) for img in base_images] 
            for key in qualifier_masks })

base_image_qalifiers
```

#### 5.7 сформований набір символів, що потрібно розпізнати (у вигляді табл. 4.3)

```python
# – {x1–x4} – символи відповідають символам А1, А3, А7, А9 відповідно з підкресленням;
# – {x5–x8} – символи відповідають символам А2, А4, А6, А8 відповідно з виділенням курсивом;
# – {x9–x11} – символи відповідають символам А2, А5, А6 відповідно з виділенням жирним;
# – {x12–x15} – символи відповідають символам А1, А4, А7, А10 відповідно з використанням шрифту «Verdana»;
# – {x16} – символ відповідає символу А10, який задано числом, з додаванням одиниці (наприклад, для А10 = «2», x16 = 2 + 1 = «3»). Якщо при додаванні одиниці число стає рівним 10 (що відповідає А10 = «9»), то x16 прийняти рівним «0». 

x1_4 = list(gen_images([letters[0], letters[2], letters[6], letters[8]], fontName = "times.ttf", underline=True))
x5_8 = list(gen_images([letters[1], letters[3], letters[5], letters[7]], fontName = "timesi.ttf"))
x9_11 = list(gen_images([letters[1], letters[4], letters[5]], fontName = "timesbd.ttf"))
x12_15 = list(gen_images([letters[0], letters[3], letters[6], letters[9]], fontName = "verdana.ttf"))
x16 = list(gen_images(['8'], fontName = "times.ttf"))

test_letters = [letters[i] for i in [0, 2, 6, 8, 1, 3, 5, 7, 1, 4, 5, 0, 3, 6, 9]] + ['8']
test_images = x1_4 + x5_8 + x9_11 + x12_15 + x16

test_base_letters = [letters[i] for i in [0, 2, 6, 8, 1, 3, 5, 7, 1, 4, 5, 0, 3, 6, 9, 9]]
test_base_images = list(gen_images(test_base_letters, fontName = "times.ttf"))

Table(["#", "C", "P", "C(Зобр)", "P(Зобр)"], [
    [
         i+1, 
         test_base_letters[i], 
         test_letters[i], 
         bw_matrix_from_image(test_base_images[i]), 
         bw_matrix_from_image(img)
     ] for (i, img) in enumerate(test_images)
])
```

#### 5.8 значення ознак для символів набору, що потрібно розпізнати (у вигляді табл. 4.8);

```python
base_image_qalifiers = pd.DataFrame({
            key:[calc_sign(img, qualifier_masks[key]) for img in test_images] 
            for key in qualifier_masks }).T

base_image_qalifiers
```

#### 5.9 перелік символів (альтернатив), що входять до множини Парето та Слейтера для кожного символу набору, що розпізнається (16-ть таблиць у вигляді табл. 4.15);

```python
for image_index, image in enumerate(test_images):
    comparision_result = calc_image_comparision(base_image_qalifiers.T, image, qualifier_masks)
    print(str(image_index + 1) + ": " + test_letters[image_index])
    display(comparision_result)
```
